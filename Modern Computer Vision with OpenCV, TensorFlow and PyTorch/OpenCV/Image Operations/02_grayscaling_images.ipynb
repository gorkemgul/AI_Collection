> We're gonna define our imshow function againg, but this time we'll change it a little bit.

```python
# Define a function to show the image
def imshow(title = 'Image', image = None, size = 10):
  w, h = image.shape[0], image.shape[1]
  aspect_ratio = w / h
  plt.figure(figsize = (size * aspect_ratio, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
# Load our input image
image = cv2.imread('./images/castara.jpeg')

# Show the image with using our imshow function
imshow(title = 'Castara, Tobago',
       image = image,
       size = 8)
```

## Turn our RGB image into GrayScale

To be able to turn our image into grayscale, we need to use `cv2.cvtColor` function.

```python
# Turn our image into grayscale
grayImage = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Let's show our gray image
imshow(title = 'Gray Scale Castara, Tobago', 
       image = grayImage,
       size = 8)
```

## Comparing the shapes of the images

🔑**Note:** RGB color images have 3 dimensions, one for each primary color. But Grayscale images just have 1, which is the intensity of gray.
