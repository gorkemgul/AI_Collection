# Drawing on Images

In this kernel we're gonna learn to:
1. Drawing a Line
2. Drawing a Rectangle
3. Drawing a Circle
4. Drawing a Polygon
5. Adding Text

```python
# Import dependencies
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Let's rewrite our imshow function
def imshow(image, title = '', size = 8):
  """
  Takes an image, turns it into RGB and shows it.

  Args:
    image: an image 
    title: name of the image
    size = a number to calculate the figsize
  
  Returns:
    an image in RGB format
  """
  w,h = image.shape[0], image.shape[1]
  aspect_ratio = w / h
  plt.figure(figsize = (size * aspect_ratio, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title);
```

```python
# Create a black image using numpy
image = np.zeros((512, 512, 3), dtype = np.uint8)

# Can we make this in black and white? grayscale
image_gray = np.zeros((512, 512), np.uint8)

# Black would be the same as greyscale or color image (same for white)
imshow(image, 'Black Canvas - RGB Color')
imshow(image_gray, 'Black Canvas - Grayscale')
```

## Drawing a Line

We can draw a line by using `cv2.line` function.

`cv2.line(image, starting cordinates, ending cordinates, color, thickness)`

```python
# Draw a line over our image
cv2.line(image, (0, 0), (511, 511), (255, 127, 0), 5)
imshow(image, 'Black Canvas with Diagonal Line', size = 6)
```

> 🔑**Note:** This is an inplace operation which means it changes the input image.

## Drawing a Rectangle

We can also draw a rectangle over our image.

To draw a rectangle, we need to use `cv2.rectangle` function.

`cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)`

```python
# Let's draw a rectangle
cv2.rectangle(image, (100, 100), (300, 250), (127, 50, 127), 10)
imshow(image, 'Black Canvas With a Rectangle', size = 6)
```

> 🔑**Note:** As we've explained when drawing a line, rectangle, etc. We need to create an image again. Because these are inplace operations. Meaning that it changes our images.

```python
# Let's create black canvas again
image = np.zeros((512, 512, 3), np.uint8)

# Draw a rectangle over our recreated image
cv2.rectangle(image, (100, 100), (300, 250), (127, 50, 127), 10)
imshow(image, 'Black Canvas with a Rectangle', size = 6)
```

## Drawing a Circle

We've drawn a line and a rectangle so far. Now let's draw a circle over our image.

To do so, we need to use `cv2.circle` function.

`cv2.circle(image, center, radius, color, fill)`

```python
# Let's recreate our image
image = np.zeros((512, 512, 3), np.uint8)

# Draw a circle over our image
cv2.circle(image, (350, 350), 100, (15, 150, 50), -1)
imshow(image, 'Black Canvas with a Circle', size = 6)
```

> 🔑**Note:** If we choose a negative thickness our circle or rectangle is being filled.

## Drawing a Polygon

Let's look at how to draw polygons over our images.

To draw a polygon, we need to use `cv2.polylines` function.

`cv2.polylines(image, points, Closed?, color, thickness)`

```python
# Create our black canvas again
image = np.zeros((512, 512, 3), np.uint8)

# Define four points 
pts = np.array([[10, 50], [400, 50], [90, 200], [50, 500]], np.int32)

# Reshape our points in form required by polylines
pts = pts.reshape((-1, 1, 2))

# Draw a polygon
cv2.polylines(image, [pts], True, (0, 0 ,255), 3)
imshow(image, 'Black Canvas with Polygon', size = 6)
```

> 🔑**Note:** The reason why we use uint is it's commonly used for storing RGB values of pixels in an image.

## Adding Text 

We've seen how to draw a line, a rectange, a circle and a polygon so far. Now let's jump into adding text to our images.

We could add a text to our images by using `cv2.putText` function.

`cv2.putText(image, 'Text to Display', bottom left starting point, Font, Font Size, Color, Thickness)`

We could use these fonts below:
* FONT_HERSHEY_SIMPLEX
* FONT_HERSHEY_PLAIN
* FONT_HERHSEY_DUPLEX
* FONT_HERHSEY_COMPLEX
* FONT_HERHSEY_TRIPLEX
* FONT_HERHSEY_COMPLEX_SMALL
* FONT_HERHSEY_SCRIPT_SIMPLEX
* FONT_HERHSEY_SCRIPT_COMPLEX


```python
# Create another image
image = np.zeros((1000, 1000, 3), np.uint8)

# Define the text to demonstrate
text = 'Hello World'

# Add the text
cv2.putText(image, text, (155, 290), cv2.FONT_HERSHEY_SIMPLEX, 3, (40, 200, 0), 4)
imshow(image, 'Black Canvas with a Text', size = 6)
```
