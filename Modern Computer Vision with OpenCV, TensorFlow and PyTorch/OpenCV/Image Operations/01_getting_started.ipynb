## Loading Images

```python
# Load an image using 'imread' specifying the path to image
image = cv2.imread('./images/flowers.jpeg')
```

## Displaying Images

```python
# Show the image with matplotlib
import matplotlib.pyplot as plt
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

> 🔑 **Note:** OpenCV doesn't treat RGB in order. It treated as BGR, so we have to reconvicted back to RGB by using `cv2.cvtColor()` function. 

Let's create a simple function to make displaying our images simpler and easier

```python
def imshow(title = '', image = None):
  """
  It takes an image and turns it from BGR to RGB
  """
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
# Let's test our function out
imshow("Displaying Our First Image", image)
```

```python
# Display our image without using 'cv2.cvtColor()' function
plt.imshow(image);  
```

## Saving Images

```python
# Save the image
cv2.imwrite('output.jpg', image)
```

```python
# We can save our image as a PNG (Portable Network Graphics)
# PNG is a lossless bitmap image format
cv2.imwrite('output.png', image)
```
